```python
# Imputing missing values in review ratings column and  transforming into median values by category
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# renaming columns according to snake casing for better readability and documentation
df.columns = df.columns.str.strip()   # remove extra spaces from start/end
df.columns = df.columns.str.lower()   # lowercase
df.columns = df.columns.str.replace(' ', '_') 
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})


# df.columns = (
#    df.columns
#   .str.strip()
#   .str.lower()
#   .str.replace(r'\s+', '_', regex=True))
```

```python
# create new age column
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_groups']= pd.qcut(df['age'], q=4, labels = labels)



# bins = [18, 30, 45, 60, 100]
# labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
# df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# checks whether discount_applied and promo_code_used are identical for every row.

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# dropping promo code used column.

df= df.drop('promo_code_used', axis =1)
```

```python
 
df['purchase_year']     = df['purchase_date'].dt.year
# → gives: 2022, 2023, 2024
 
df['purchase_month']    = df['purchase_date'].dt.month
# → gives: 1, 2, 3, ... 12  (as integers)
 
df['purchase_month_name'] = df['purchase_date'].dt.strftime('%b')
# → gives: 'Jan', 'Feb', 'Mar' ... 'Dec'  (short text names)
 
df['purchase_quarter']  = df['purchase_date'].dt.quarter
# → gives: 1, 2, 3, or 4  (Q1 = Jan-Mar, Q2 = Apr-Jun...)
 
df['purchase_week']     = df['purchase_date'].dt.isocalendar().week.astype(int)
# → gives: 1-52  (week number of the year)
 
df['purchase_day_name'] = df['purchase_date'].dt.day_name()
# → gives: 'Monday', 'Tuesday' ... 'Sunday'
 
df['is_weekend']        = df['purchase_day_name'].isin(['Saturday','Sunday'])
# → gives: True or False  (useful for weekend vs weekday analysis)
 

```

```python
# Define which months belong to each season
season_month_map = {
    'Winter': [12, 1, 2],
    'Spring': [3, 4, 5],
    'Summer': [6, 7, 8],
    'Fall':   [9, 10, 11]
}
```

```python
# Create a helper function that returns True if date matches season
df['season_date_match']= df.apply(
    lambda row: row['purchase_month'] in season_month_map.get(row['season'] , []), axis=1)
```

```python
# Count mismatches
mismatches = df[~df['season_date_match']]
print(f'Total mismatches: {len(mismatches)}')  

 
```

```python
# Drop the helper column (no longer needed)
df = df.drop('season_date_match', axis=1)
```

```python
# Days since last purchases (Recency)

reference_date= df['purchase_date'].max()
df['days_since_purchase']= (reference_date - df['purchase_date']).dt.days
```

```python
# Spend Segment (Customer grouping)

df['spend_segment'] = pd.cut(
    df['purchase_amount'],
    bins=[0, 30, 60, 80, 100],
    labels=['Low ($0-30)', 'Medium ($31-60)', 'High ($61-80)', 'Premium ($81-100)']
)
```

```python
# Estimated Purchase per year

df['est_purchases_per_year']=(365/df['purchase_frequency_days']).round(1)
```

```python
# Estimated Annual Value

df['estimated_annual_value'] = (
    df['purchase_amount'] * df['est_purchases_per_year']
).round(2)
```
